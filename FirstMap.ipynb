```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
```

```python
data = np.load('FirstMap.npz')
az = data['az']
el = data['el']
pwr = data['pwr']
```

```python
plt.plot(az,el,'.')
plt.show()
```

```python
plt.plot(pwr)
plt.show()
```

```python
eli = np.linspace(az.min(),az.max(),20)
azi = np.linspace(el.min(),el.max(),20)
# grid the data.
zi = griddata((az, el), pwr, (eli[None,:], azi[:,None]), method='nearest')
# contour the gridded data, plotting dots at the randomly spaced data points.
#plt.imshow(zi,aspect='auto',cmap=plt.cm.jet)
CS = plt.contour(eli,azi,zi,5,linewidths=0.5,colors='w')
CS = plt.contourf(eli,azi,zi,10,cmap=plt.cm.jet)
plt.axis('equal')
plt.show()
```

```python
plt.imshow(zi-32.,aspect='auto',cmap=plt.cm.jet)
plt.colorbar()
plt.axis('equal')
plt.show()
```
